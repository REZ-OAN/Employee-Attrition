Now we have to convert the categorical value to numeric value

Here Two types of categorical value can be there

1. categorical data with their ordinal relationship ( has a meaning of their order)

2. categorical data without their ordinal relationship (do not have meaning of their order)

Now we have to findout the type-1 and type-2 categorical data in our dataset

for type-1 we  used the label encoder to covert them into numeric value

for type-2 wel used the one-hot encoder to covert them into numeric value

type-1 categorical data in our dataset

1.Attrition

2.BussinessTravel

3.Department

4.EducationField

5.JobRole

6.MaritalStatus

7.OverTime

type-2 categorical data in our dataset

1.Gender

```python
#It is used for data preprocessing tasks like scaling, encoding
from sklearn import preprocessing
```

```python
# Label Encoding
label_encoder = preprocessing.LabelEncoder()

df['attrition'] = label_encoder.fit_transform(df['Attrition'])
## printing the label mapping
label_mapping = dict(zip(df['attrition'], df['Attrition']))
## print in sorted order
print('For Attrition Column')
print(dict(sorted(label_mapping.items(), key=lambda item: item[1])));

df['businessTravel'] = label_encoder.fit_transform(df['BusinessTravel'])
## printing the label mapping
label_mapping = dict(zip(df['businessTravel'], df['BusinessTravel']))
## print in sorted order
print('For BusinessTravel Column')
print(dict(sorted(label_mapping.items(), key=lambda item: item[1])));

df['department'] = label_encoder.fit_transform(df['Department'])
## printing the label mapping
label_mapping = dict(zip(df['department'], df['Department']))
## print in sorted order
print('For Department Column')
print(dict(sorted(label_mapping.items(), key=lambda item: item[1])));

df['educationField'] = label_encoder.fit_transform(df['EducationField'])
## printing the label mapping
label_mapping = dict(zip(df['educationField'], df['EducationField']))
## print in sorted order
print('For EducationField Column')
print(dict(sorted(label_mapping.items(), key=lambda item: item[1])));

df['jobRole'] = label_encoder.fit_transform(df['JobRole'])
## printing the label mapping
label_mapping = dict(zip(df['jobRole'], df['JobRole']))
## print in sorted order
print('For JobRole Column')
print(dict(sorted(label_mapping.items(), key=lambda item: item[1])));

df['maritalStatus'] = label_encoder.fit_transform(df['MaritalStatus'])
## printing the label mapping
label_mapping = dict(zip(df['maritalStatus'], df['MaritalStatus']))
## print in sorted order
print('For MaritalStatus Column')
print(dict(sorted(label_mapping.items(), key=lambda item: item[1])));

df['overTime'] = label_encoder.fit_transform(df['OverTime'])
## printing the label mapping
label_mapping = dict(zip(df['overTime'], df['OverTime']))
## print in sorted order
print('For OverTime Column')
print(dict(sorted(label_mapping.items(), key=lambda item: item[1])));

df['over18'] = label_encoder.fit_transform(df['Over18'])
## printing the label mapping
label_mapping = dict(zip(df['over18'], df['Over18']))
## print in sorted order
print('For Over18 Column')
print(dict(sorted(label_mapping.items(), key=lambda item: item[1])));

 #  Removing the  unnecessary and same meaning columns
df.drop(columns=['Attrition','BusinessTravel', 'Department', 'EducationField','JobRole', 'MaritalStatus','OverTime','Over18'],inplace=True)

# overview of the data after label encoding and removing the unnecessary columns
df.head(10)
```

```python

from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn import tree
## Models
from sklearn.linear_model import LogisticRegression

from sklearn.tree import DecisionTreeClassifier


from sklearn.ensemble import RandomForestClassifier


from sklearn.neighbors import KNeighborsClassifier
```

**Applying Logistic Regression Algorithm**

```python

LR = LogisticRegression()
LR.fit(X_train,y_train)
```

```python
LR_score = LR.score(X_train, y_train)
LR_test = LR.score(X_test, y_test)

#testing model

y_pred = LR.predict(X_test)

#evaluation

cm = confusion_matrix(y_test,y_pred)
print(classification_report(y_test, y_pred))
models_scores['LR_MODEL'] = accuracy_score(y_test,y_pred)
```

```python
# Visualizing Confusion Matrix
plt.figure(figsize = (8, 5))
sns.heatmap(cm, cmap = 'Blues', annot = True, fmt = 'd', linewidths = 5, cbar = False, annot_kws = {'fontsize': 15},
            yticklabels = ['No stroke', 'Stroke'], xticklabels = ['Predicted no stroke', 'Predicted stroke'])
plt.yticks(rotation = 0)
plt.show()
```

```python
## intialize the Decision Tree model
DT_model=DecisionTreeClassifier(max_depth=3, random_state=1)
```

```python
## train the decision tree model
DT_model.fit(X_train, y_train)
## ploting the deth of Decision tree model
plt.figure(figsize=(20, 12))
tree.plot_tree(DT_model, filled=True)
plt.show()
## testing the decision tree model
DT_ypred = DT_model.predict(X_test)
## printing accuracy of decision tree model
print("Test Accuracy", accuracy_score(y_test, DT_ypred))
print("Test Classification Report\n", classification_report(y_test,DT_ypred))
## illustrating the confusion matrix of decition tree model prediction
mat=confusion_matrix(y_test,DT_ypred)
models_scores['DT_MODEL']=accuracy_score(y_test,DT_ypred)
sns.heatmap(mat,annot=True, cbar=False,cmap="Blues")
plt.show()
```

```python
## initializing the Random Forest Classifier
RFC_model = RandomForestClassifier(n_estimators = 100)
```

```python

## Train the Random Forest model  with train dataset
RFC_model.fit(X_train, y_train)
## testing the Random Forest model
RFC_ypred = RFC_model.predict(X_test)
## printing the Accuracy of the model
print("Test Accuracy", accuracy_score(y_test, RFC_ypred))
print("Test Classification Report\n", classification_report(y_test,RFC_ypred))
## illustrating the confusion matrix for the prediction of the model
mat=confusion_matrix(y_test,RFC_ypred)
models_scores['RF_MODEL']=accuracy_score(y_test,RFC_ypred)
sns.heatmap(mat,annot=True, cbar=False,cmap="Blues")
plt.show()
```

```python
#### initializing the KNN Classifier
KNN_model = KNeighborsClassifier(n_neighbors=5, metric='euclidean')
```

```python
## Train the KNN model  with train dataset
KNN_model.fit(X_train, y_train)
## testing the KNN model with test dataset
X_test_contigious = np.ascontiguousarray(X_test)
KNN_ypred = KNN_model.predict(X_test_contigious)
## printing the accuracy of the prediction of KNN model
print("Test Accuracy", accuracy_score(y_test, KNN_ypred))
print("Test Classification Report\n", classification_report(y_test,KNN_ypred))
## illustrating the confusion matrix of KNN model prediction
mat=confusion_matrix(y_test,KNN_ypred)
models_scores['KNN_MODEL']= accuracy_score(y_test,KNN_ypred)
sns.heatmap(mat,annot=True, cbar=False,cmap="Blues")
plt.show()
```
