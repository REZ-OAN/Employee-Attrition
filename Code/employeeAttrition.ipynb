```python
# visualizing the data Marital Status Based On Employee Attrition
custom_palette = {'Yes': 'blue', 'No': 'lightblue'}
sns.countplot(x='MaritalStatus', hue='Attrition', data=df,palette=custom_palette)
fig = plt.gcf()
fig.set_size_inches(20,8)
plt.title('Marital Status based on Employee Attrition')
plt.xlabel('Marital Status')
plt.ylabel('Count')
plt.show()
```

As Age Column has many unique values,

we have visualize Age Column Based On Attrition Column,

in a histogram plot

```python
## Visualizing Age based on Employee Attrition
plt.figure(figsize=(20, 8))
sns.histplot(data=df, x="Age", hue="Attrition", element="bars", multiple="stack", fill=True,palette="rocket")
plt.title('Attrition Status by Age', fontsize=15)
plt.xlim([df['Age'].min(), df['Age'].max()]) 
plt.legend(title='Attrition', labels=['No', 'Yes'])
plt.show()
```
